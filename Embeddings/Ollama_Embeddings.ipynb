```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings = (
    OllamaEmbeddings(model="llama3.2") ## By default it uses gammma2
)
```

### After executing this I saw an error that says that this method is now deprecated and now a different library will be used for embedding so using that one.

```python
from langchain_ollama import OllamaEmbeddings
embeddings = (
    OllamaEmbeddings(model="llama3.2")
)
```

```python
embeddings
```

```python
result = embeddings.embed_documents([
    "My first name is Akshit",
    "My last name is Balodhi"
])
```

```python
result
```

```python
result2 = embeddings.embed_query("What is my last name?")
print(result2)
```
