#### Sentence Transformers

Sentence Transformers is a state of the art python framework for sentence, text and image embeddings. It is widely used in the industry because it converts the text into a dense vector of depth 384 which is very small and optimized.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
result = embeddings.embed_query("Banana is a fruit with no seeds.")
result
```

```python
print(len(result))
```

```python
documents = embeddings.embed_documents(["Watermelon is a fruit with many seeds.",
                                        "Banana is a fruit with no seeds.",
                                       "Avacado is a fruit with one seed."])
documents
```
